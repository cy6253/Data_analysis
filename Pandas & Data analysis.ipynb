# **5. 날짜 데이터 처리**

```python
df1 = pd.read_csv('Data01.csv')
df1.shape
```

# **7. 엑셀 데이터 처리**

점포 별 판매량 추이를 그래프로 표현

상품의 종류를 나누어 표현

가장 많이 팔린 점포 확인

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family = 'Malgun Gothic')
```

```python
df1 = pd.read_csv('Data09.csv', encoding = 'cp949')
df1.shape
```

```python
df1.head(5)
```

```python
pd.DataFrame(df1.drop(columns=['순번', '상품코드']).stack()).reset_index().iloc[640:650]
```

```python
df1.drop(columns=['순번', '상품코드']).set_index('상품명') # 상품명도 stack이 같이 되는것을 방지하기 위해 상품명을 인덱스로 사용
```

```python
stack_data = df1.drop(columns=['순번', '상품코드']).set_index('상품명').stack()
stack_data
```

```python
df2 = pd.DataFrame(stack_data).reset_index()
df2.head(10)
```

```python
df2['상품명'].isnull()
```

```python
cond1 = df2['상품명'].isnull()
cond2 = df2['상품명']=='합 계'
cond3 = df2['level_1']==' 합계'
df3 = df2.loc[~cond1 & ~cond2 & ~cond3]
df4 = df3.rename(columns={'level_1':'지점명', 0:'판매량'})
df4
```

```python
def func1(row):
  if 'CAT' in row:
    return 'C type'
  elif 'STL' in row:
    return 'S type'
  elif 'BW' in row:
    return 'B type'
  else:
    return 'Non type'
```

```python
df4['제품군'] = df4['상품명'].apply(func1)
```

```python
df4
```

```python
df4['제품군'].value_counts()
```

```python
cond1 = df4['상품명']!='Non type'
df5 = df4.loc[cond1]
```

```python
df5.info() # 판매량이 object
```

```python
def func2(row):
  if len(row)>=4:
    result = row.split(',') # remove 사용 가능
    return result[0] + result[1]
  else:
    return row

df5['판매량(int)'] = df5['판매량'].apply(func2).astype('int')
```

```python
df5.info()
```

```python
df6 = pd.pivot_table(data = df5, index = ['지점명', '제품군'], values = '판매량(int)', aggfunc=sum).reset_index()
```

```python
df7 = df6.sort_values(by = '판매량(int)', ascending=False).head(200)
```
